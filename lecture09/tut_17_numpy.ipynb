### Array Operations

In python list, we cannot simply add the elements from two list.\
For example,

```python
# list add up
a = [x for x in range(int(1e5))]   # [0,1,2,3,4....]
b = [x+1 for x in range(int(1e5))] # [1,2,3,4,5,...]
print(a+b)   # we want [1,3,5,7,9, ...], but we don't get it! Please see in the following cell codes.
print(len(a + b))
```

you need to write a funtion to do this:

```python
def vector_add(a,b):
    return [v1+v2 for v1,v2 in zip(a,b)]
```

```python
c = vector_add(a,b)
print(c)
```

remember that the python loop of a list is very slow

```python
%timeit vector_add(a,b)
```

We could do this calculation much better with numpy

```python
np_a = np.array(a)
np_b = np.array(b)
np_c = np_a + np_b
print(np_c)
```

```python
%timeit (np_a + np_b)
```

It's roughly x65 speed up.\
Do you get a similiar speed up?

```python
# more array operations

np.add(np_a,np_b) # same as np_a + np_b
np.subtract(np_a, np_b) # same as np_a - np_b

# we could also do
10 * np_a    # times 10 in each element
5 / np_a     # 5 / np_a[:]
np_a / 5     # np_a[:] / 5
np.sin(np_a) # sin
np.cos(np_a) # cos 
```

However, what if you want to join two arrays as python list1 + list2?\
We could do this with `np.concatenate()`.


```python
np_a = np.ones(4)
np_b = np.zeros(5)
np_c = np.concatenate((np_a,np_b))
print(np_c)
```

`np.concatenate()` can be applied to multi-D arrays as well, but you need to assign an axis

```python
np_a = np.array([[1,2],[3,4]])
np_b = np.array([[5,6],[7,8]])
print(np_a.shape, np_b.shape)
np_c1 = np.concatenate((np_a,np_b)) # default: axis = 0
print(np_c1)
np_c2 = np.concatenate((np_a,np_b), axis = 1) # along rows
print(np_c2)
```

Stacking is same as concatenation, the only difference is that stacking is done along a new axis.\
Stacking two 1D arrays will be a 2D array.

```python
np_a = np.ones(5)
np_b = np.zeros(5)
np_c = np.stack((np_a,np_b)) # default: axis = 0
print(np_c)
np_c = np.stack((np_a,np_b), axis=1) # default: axis = 0
print(np_c)
```

There are also `hstack` and `vstack` functions

```python
np_c = np.hstack((np_a, np_b)) # stack along rows 
print(np_c)
```

```python
np_c = np.vstack((np_a, np_b)) # stack along columns
print(np_c)
```

```python
np_c = np.dstack((np_a, np_b)) # stack along depth
print(np_c)
```

## NumPy MeshGrid

When describing physical world, it is often that we want to describe the physical locations in Cartesian coordinates.\
This can be achived with `np.meshgrid()`.

```python
# cosider 30 x 40 girds within a domain [xmin:xmax, ymin:ymax]
nx = 30
ny = 40

xmin = -1.0
xmax =  2.0
ymin = 0
ymax = 4

# first define the x-coordinates
x = np.linspace(xmin, xmax, nx, endpoint=False)
y = np.linspace(ymin, ymax, ny, endpoint=False)

# the 2D grids can be defined by
# xx, yy = np.meshgrid(x, y, indexing='xy') # indexing='xy' is the default
# for i in range(nx):
#     for j in range(ny):
#         pass
        #print(xx[j,i], yy[j,i])

# note that the above index xx[j,i] is different with what we used in physics
# it is better to use indexing='ij' when describing physcical coordinates.

xx, yy = np.meshgrid(x,y, indexing='ij')
for i in range(nx):
    for j in range(ny):
        pass
        # print(xx[i,j], yy[i,j])
```

Once we have the 2D grids `xx` and `yy`, we could use it to define physcail quantities in this grid.

```python
zz = xx*(yy-2)**2 # define a field `z = x * (y-2)**2`
```

We could use `matplotlib` to visualize the quantity.

```python
try:
    plt.figure(1)
    plt.imshow(zz.T, origin='lower', extent=[xmin, xmax, ymin, ymax])
    plt.xlabel("X")
    plt.ylabel("Y")
except:
    print("Error: cannot use matplotlib.")
```

## NumPy Exercise 2: MeshGrid

Calculate the gravitational potential of a binary stars.
* Star 1 has a mass, m1 = 2 solar mass, located at (x,y) = (-5 AU, 0)
* Star 2 has a mass, m2 = 5 solar mass, located at (x,y) = (2 AU, 0)
* Consider a 500 x 500 grids in 2D Cartesain coordinates
* The domain includes from - 15 AU to 15 AU in each dimension. 


constants:\
    Solar mass: 1.989e33 [gram]\
    Astronomy Unit (AU): 1.5e13 [cm]\
    Gravitational constant, G: 6.67e-8 [cgs]

```python
def potential(G, m, d):
  return -G*m/d
```

```python
# TODO

solar_mass = 1.989e33
AU = 1.5e13
G = 6.67e-8
m1 = 2*solar_mass
pos1_x, pos1_y = -5*AU, 0
m2 = 5*solar_mass
pos2_x, pos2_y = 2*AU, 0

nx = 500
ny = 500

xmin = -15
xmax =  15
ymin = -15
ymax =  15

x = np.linspace(xmin, xmax, nx, endpoint=False)
y = np.linspace(ymin, ymax, ny, endpoint=False)

xx, yy = np.meshgrid(x, y, indexing='ij')
dd1 = np.sqrt((xx*AU - pos1_x)**2 + (yy*AU - pos1_y)**2)
dd2 = np.sqrt((xx*AU - pos2_x)**2 + (yy*AU - pos2_y)**2)
zz1 = np.log10(np.abs(potential(G, m1, dd1)))
zz2 = np.log10(np.abs(potential(G, m2, dd2)))

plt.figure(1)
plt.imshow((zz1+zz2).T,
           origin='lower', 
           extent=[xmin, xmax, ymin, ymax])
# plt.imshow(zz2.T, origin='lower', extent=[xmin, xmax, ymin, ymax])
plt.xlabel("X [AU]")
plt.ylabel("Y [AU]")
plt.colorbar()
```

## Numerical Derivatives

* derivatives with `np.diff()` https://numpy.org/doc/stable/reference/generated/numpy.diff.html

```python
# Example 1:
# let's start from a simple x
x = np.arange(10)
print(f" x = {x}")
print(len(x))
# np.diff will returns x[i+1] - x[i]
dx = np.diff(x)
print(f"dx = {dx}")
print(len(dx))
```

note that `np.diff()` will reuturns a numpy arrays with N-1 elements.

```python
# Example 2:
# a slightly more complicate function
f = x**2
print(f" f(x) = {f}")
print(len(x))
#
df = np.diff(f)
print(f"df(x) = {df}")
print(len(dx))
```

```python
# Example 3:
# 
x = np.linspace(0,2*np.pi,30)

# f(x) = sin(x)
fx   = np.sin(x)    # N   elements
df   = np.diff(fx)  # N-1 elements
dx   = np.diff(x)   # N-1 elements
dfdx = df/dx        # N-1 elements
sol  = np.cos(x)    # True solutions

plt.figure(1)
plt.plot(x,fx,label="f(x) = sin(x)")
plt.plot(x[0:-1],dfdx,label="method 1")
plt.plot(x[1:],dfdx,label="method 2")
plt.plot(x,sol, label="sol. cos(x)")
plt.legend(loc='best')
plt.xlabel('x')
```

### Gradient 

`np.gradient` is very similar to `np.diff` but will return the same length of the input.

REF: https://numpy.org/doc/stable/reference/generated/numpy.gradient.html

```python
# x from 0 to 2pi
x = np.linspace(0,2*np.pi,10)

# use np.gradient
dx1 = np.gradient(x)

# use np.diff
dx2 = np.diff(x)
print(dx1)
print(dx2)
```

```python
# x from 0 to 2pi
x = np.linspace(0,2*np.pi,10)

# f(x) = sin(x)
fx   = np.sin(x)
sol  = np.cos(x)
dfdx = np.gradient(fx,x)

plt.figure(1)
plt.plot(x,fx,label="f(x) = sin(x)")
plt.plot(x,dfdx,label="np.gradient")
plt.plot(x,sol, label="sol. cos(x)")
plt.legend(loc='best')
plt.xlabel('x')
```

You could see that the result with `np.gradient` is much better than the result with `np.diff`.\
It is becasue the "method 1" and "method 2" we used in the above example are first-order.\
`np.gradient` is second-order (except the edge).

To use second-order on the edge, `edge_order=2` in the argument must be passed.

```python
dfdx2 = np.gradient(fx,x, edge_order=2)
plt.figure(1)
plt.plot(x,fx,label="f(x) = sin(x)")
plt.plot(x,dfdx, label="np.gradient (eo 1)")
plt.plot(x,dfdx2,label="np.gradient (eo 2)")
plt.plot(x,sol, label="sol. cos(x)")
plt.legend(loc='best')
plt.xlabel('x')
# note that this example is not very good. We got overshooting on the edges.
```
